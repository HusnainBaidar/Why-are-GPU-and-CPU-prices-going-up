```python
docs = []
stemmer = WordNetLemmatizer()
for sen in range (0,len (data_list)):
    #special char
    doc = re.sub(r'\W', ' ',str(data_list[sen]))
    
    #punctuation
    doc = re.sub(re.escape(string.punctuation),'',doc)
    
    #no singel char
    doc = re.sub(r'\s+[a-zA-Z]\s+', ' ',doc)
    
    #no single char from start
    doc = re.sub(r'\^[a-zA-Z]\s+', ' ',doc)
    
    #multiple space with single space
    doc = re.sub(r'\s+', ' ',doc,flags=re.I)
    
    #removing prefix of b
    doc = re.sub(r'^b\s+',' ',doc)
    
    #converting to lowercase
    doc = doc.lower()
    
    # Print string after removing tags
    doc = re.compile(r'<[^>]+>').sub('', doc)
    
    doc = re.sub("p", "", doc)
    doc = re.sub("<p>", "", doc)
    doc = re.sub("</p>", "", doc)
    doc = re.sub("href", "", doc)
    doc = re.sub(" r ", "", doc)
    doc = re.sub("class", "", doc)
    doc = re.sub("www", "", doc)
    doc = re.sub("htts", "", doc)
    doc = re.sub("1...", "", doc)
    doc = re.sub("com", "", doc)
    doc = re.sub(r'[0-9]+', '', doc)
    doc = re.sub(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', '', doc)

    #limitization
    doc = doc.split()
    doc = [stemmer.lemmatize(word) for word in doc]
    doc = ' '.join(doc)
    doc = [word for word in doc.split() if word not in stopwords.words('english')]
    doc = ' '.join(doc)
    
    docs.append(doc)
```
